# Hypothesis Test for Average Wait Time

A coffee shop claims that their average wait time for customers is less than 5 minutes. 
To test this claim, a sample of 40 customers is taken, and their wait times are recorded. 
The sample mean wait time is found to be 4.6 minutes with a standard deviation of 0.8 minutes. Perform a hypothesis
test at a significance level of 0.05 and determine whether 
there is enough evidence to support the coffee shop's claim.

# Given Data

- Sample mean$(\(\bar{x}\))$: 4.6 minutes
- Sample standard deviation (\(s\)): 0.8 minutes
- Sample size (\(n\)): 40
- Population mean under the null hypothesis$(\(\mu_0\))$: 5 minutes
- Significance level$(\(\alpha\))$: 0.05

## Hypotheses

- Null hypothesis (\(H_0\)): $\(\mu \leq 5\)$
- Alternative hypothesis (\(H_a\)): $\(\mu > 5\)$

## Test Statistic

The formula for the t-statistic is:

$$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$$

Plugging in the values:

$$t = \frac{4.6 - 5}{0.8 / \sqrt{40}}$$

## Critical Value and p-value

For a one-tailed test at $\(\alpha = 0.05\)$ and \(df = n - 1\):

- Degrees of freedom (\(df\)): 39
- Critical t-value: $\(t_{\alpha, df}\)$
- p-value:\(1 - $\text{CDF}(t)\)$

## Conclusion

If the t-statistic is greater than the critical t-value and the p-value is less than \(\alpha\), we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.


#One sample t-test - right tailed 

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

We will define the given data, including the sample mean, sample standard deviation, sample size, population mean under the null hypothesis, and the significance level.

```python
# Given data
sample_mean = 4.6
sample_std_dev = 0.8
sample_size = 40
population_mean = 5
alpha = 0.05
```

```python
# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
t_statistic.round(4)
```

```python
# Degrees of freedom

df = sample_size - 1

# Calculate the p-value (lower tail)
p_value_lower = stats.t.cdf(t_statistic, df)

# Determine the critical t-value for one-tailed test (lower tail)
critical_t_lower = stats.t.ppf(alpha, df)

p_value_lower, critical_t_lower
```

```python
# Calculate the p-value for the upper tail
p_value_upper = 1 - stats.t.cdf(t_statistic, df)

# Determine the critical t-value for one-tailed test (upper tail)
critical_t_upper = stats.t.ppf(1 - alpha, df)

p_value_upper, critical_t_upper
```

```python
if t_statistic > critical_t_upper and p_value_upper < alpha:
    print("Reject the null hypothesis: There is enough evidence to support the claim that the average wait time is more than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to support the claim that the average wait time is more than 5 minutes.")
```
