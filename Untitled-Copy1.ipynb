1. Z-score : usefull when data is normal distribution. identify extreme values in normal distribution
2. IQR : usefull when data is not normal distribution or if data is skewed.
3. Capping : mitifate outlier effect without removing
4. Clipping : removing extreme values.

```python
# function to identify if a distribution is normal or skewed
from scipy.stats import skew, zscore

def identify_distribution(column):
    skewness = skew(column)
    if abs(skewness) < 0.5:
        return 'normal'
    else:
        return 'skewed'

distribution_types = df[numerical_features].apply(identify_distribution)
```

```python
# keep copy of data
df_numerical = df[numerical_features].copy()

normal_features = distribution_types[distribution_types == 'normal'].index.to_list()
skewed_features = distribution_types[distribution_types == 'skewed'].index.to_list()
```

```python

# Define the IQR threshold multiplier
iqr_multiplier = 1.5

# Handle outliers for each skewed feature
for col in skewed_features:
    # Calculate Q1 and Q3 for the feature
    Q1 = df_numerical[col].quantile(0.25)
    Q3 = df_numerical[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - iqr_multiplier * IQR
    upper_bound = Q3 + iqr_multiplier * IQR
    
    # Identify outliers based on IQR bounds
    outliers = (df_numerical[col] < lower_bound) | (df_numerical[col] > upper_bound)
    
    # Replace outliers with the median value
    # median_value = df[col].median()
    # df_numerical_after_outlier_removal.loc[outliers, col] = median_value
    
    # Replace outliers with the IQR bounds instead
    df_numerical_after_outlier_removal.loc[df_numerical_after_outlier_removal[col] < lower_bound, col] = lower_bound
    df_numerical_after_outlier_removal.loc[df_numerical_after_outlier_removal[col] > upper_bound, col] = upper_bound
```

```python
##########
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Function to transform positively skewed features
def transform_positive_skew(df1, skewed_features):
    """
    Apply log(1+x) transformation to positively skewed features.
    """
    transformed_features = []
    for feature in skewed_features.index:
        try:
            # Apply log(1+x) transformation
            df1[f'log_{feature}'] = np.log1p(df1[feature])
            transformed_features.append(f'log_{feature}')
        except Exception as e:
            print(f"Error transforming {feature}: {e}")
    return df1, transformed_features

# Function to transform negatively skewed features
def transform_negative_skew(df1, skewed_features):
    """
    Apply exponential transformation to negatively skewed features.
    """
    transformed_features = []
    for feature in skewed_features.index:
        try:
            # Apply exponential transformation
            df1[f'exp_{feature}'] = np.exp(df1[feature])
            transformed_features.append(f'exp_{feature}')
        except Exception as e:
            print(f"Error transforming {feature}: {e}")
    return df1, transformed_features

# Example usage
# Calculate skewness for all numerical features
skewness = df1[numerical_features].skew()

# Filter positively skewed features (skewness > 1)
positively_skewed_features = skewness[skewness > 1].sort_values(ascending=False)

# Filter negatively skewed features (skewness < -1)
negatively_skewed_features = skewness[skewness < -1].sort_values(ascending=True)

# Transform positively skewed features
df1, pos_transformed_features = transform_positive_skew(df1, positively_skewed_features)

# Transform negatively skewed features
df1, neg_transformed_features = transform_negative_skew(df1, negatively_skewed_features)

# Combine transformed features with the target variable `shares`
transformed_features_df1 = df1[pos_transformed_features + neg_transformed_features + ['shares']]

# Display the new DataFrame
print("Transformed Features DataFrame:")
print(transformed_features_df1.head())

# Plot histograms for positively transformed features
for feature in pos_transformed_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(df1[feature], kde=True, bins=30, color='green')
    plt.title(f'Histogram of Log-Transformed {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Plot histograms for negatively transformed features
for feature in neg_transformed_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(df1[feature], kde=True, bins=30, color='blue')
    plt.title(f'Histogram of Exponentially Transformed {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```
